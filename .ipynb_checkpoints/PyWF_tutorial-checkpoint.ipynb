# Notebook to make nice plots of the comparison between my WFs and LAL

```python
import numpy as np
import time
import sys
import os
import pandas as pd
import scipy.integrate as igt
import h5py
import matplotlib.pyplot as plt
from matplotlib import rc
rc('text', usetex=True)

import PyWF 
import WFutils as utils
```

```python
from astropy.cosmology import FlatLambdaCDM, Planck18

zGridGLOB = np.logspace(start=-10, stop=3, base=10, num=1000)
dLGridGLOB = Planck18.luminosity_distance(zGridGLOB).value
dcomGridGLOB = Planck18.comoving_distance(zGridGLOB).value
HGridGLOB = Planck18.H(zGridGLOB).value

from scipy import interpolate

dcom70 = interpolate.interp1d(zGridGLOB, dcomGridGLOB, kind='cubic', bounds_error=False, fill_value=(0, np.NaN), assume_sorted=True)
dL70 = interpolate.interp1d(zGridGLOB, dLGridGLOB, kind='cubic', bounds_error=False, fill_value=(0 ,np.NaN), assume_sorted=True)
H70 = interpolate.interp1d(zGridGLOB, HGridGLOB, kind='cubic', bounds_error=False, fill_value=(70 ,np.NaN), assume_sorted=True)

def dL(z, H0):
    '''
    Standard luminosity distance
    '''
    return dL70(z) * (Planck18.H0.value/H0) 

def Xi(z, Xi0, n):
    return Xi0+(1-Xi0)/(1+z)**n

def dLGW(z, H0=Planck18.H0.value, Xi0=1., n=1.91):
    '''
    Modified GW luminosity distance
    '''
    # could try speeding this up by reducing number of function calls by one, copying in the expression for Xi...
    return Xi(z, Xi0, n) * dL70(z) * (Planck18.H0.value/H0)

fiducials = {'H0': 70, 'Xi0': 1, 'nPar': 1.9}

```

```python
import pycbc.waveform
import pycbc.types as pcbcdt

def m1m2_from_Mceta(Mc, eta):
    m1 = 0.5*(Mc/(eta**(3./5.)))*(1.+np.sqrt(1.-4.*eta))
    m2 = 0.5*(Mc/(eta**(3./5.)))*(1.-np.sqrt(1.-4.*eta))
    
    return m1, m2
```

## Start with IMRPhenomD

```python
zs = np.array([.2])
Lam = fiducials
events = {'Mc':np.array([30])*(1.+zs), 'dL':(dLGW(zs,Lam['H0'], Lam['Xi0'], Lam['nPar'])/1000.), 
          'theta':np.array([np.pi/4.]), 'phi':np.array([np.pi/4.]), 
         'iota':np.array([.0]), 'psi':np.array([np.pi/6.]), 'tcoal':np.array([.3]), 'eta':np.array([0.24]), 
          'Phicoal':np.array([np.pi*0.5]),
         'chi1z':np.array([0.8]), 'chi2z':np.array([-0.8]), 'Lambda1':np.array([0.]), 'Lambda2':np.array([0.])}
```

```python
events['dL']
```

```python
tmpWF = WF.IMRPhenomD()

fcut = tmpWF.fcut(**events)
fminarr = np.full(fcut.shape, 5)
fgrids = np.geomspace(fminarr, fcut, num=int(1000))

fgrids.shape
```

```python
Mtot = events['Mc']/(events['eta']**(3./5.))
m1s, m2s = m1m2_from_Mceta(events['Mc'], events['eta'])

hp, hc = pycbc.waveform.get_fd_waveform_sequence(approximant='IMRPhenomD',                                        
                                                    mass1=m1s,
                                                    mass2=m2s,
                                                    spin1z=events['chi1z'],
                                                    spin2z=events['chi2z'],
                                                    #lambda1=float(events['Lambda1']),
                                                    #lambda2=float(events['Lambda2']),
                                                    #dquad_mon1 = float(0.0),
                                                    #dquad_mon2 = float(0.0),
                                                    sample_points = pcbcdt.array.Array(fgrids[:,0]),
                                                    distance=events['dL']*1000.)

PyCBCwfAmpl = onp.array(abs(hp))
PyCBCwfPhase = onp.array(hp/abs(hp))
print(m1s, m2s)
```

```python
myampl = tmpWF.Ampl(fgrids, **events)
myphase = tmpWF.Phi(fgrids, **events)
```

```python
from matplotlib import gridspec
fig = plt.figure(figsize=(10,12))
gs = gridspec.GridSpec(3,1,height_ratios=[2,1,.9])
ax1 = plt.subplot(gs[0])
ax2 = plt.subplot(gs[1], sharex=ax1)
ax3 = plt.subplot(gs[2], sharex=ax1)
ax1.plot(fgrids, PyCBCwfAmpl, 'C1', label=r'$\texttt{LAL}$', alpha=.35, linewidth=6.)
ax1.plot(fgrids, myampl, 'C0', label=r'$\texttt{%s}$'%ourcodenameString)
ax1.set_yscale('log')
ax1.set_ylabel(r'$A_+ \, ({\rm Hz}^{-1})$',fontsize=20)
ax1.grid()
#ax2.plot(fgrids, np.exp(1j*(myphase)).real)
ax2.plot(fgrids, PyCBCwfPhase.real, 'C1', alpha=.35, linewidth=6.)
ax2.plot(fgrids[:,0], onp.cos(myphase[:,0]), 'C0',)
ax2.set_ylabel(r'$\cos(\Psi_{+})$', fontsize=20)
plt.xlabel(r'$f \, (\rm Hz)$', fontsize=20)
yticks = ax2.yaxis.get_major_ticks()
yticks[-1].label1.set_visible(False)
#ax3.plot(fgrids[:,0], (1.-(myampl[:,0]*np.cos(myphase[:,0]).real)/(onp.array(hp).real)), '.', color='C2', ms=3)
#ax3.plot(fgrids[:,0], myampl[:,0]*np.cos(myphase[:,0])-(PyCBCwfAmpl*PyCBCwfPhase.real), '.', color='C2', linewidth=.5)
ax3.plot(fgrids[:,0], abs(1.-(myampl[:,0])/(PyCBCwfAmpl)), '.', color='C2', ms=3)
#ax3.set_ylim(1e-10,1e-3)
ax3.set_ylabel(r'$\rm res$', fontsize=20)
ax1.set_xscale('log')
ax3.set_yscale('log')

plt.subplots_adjust(hspace=0.)

handles, labels = ax1.get_legend_handles_labels()
ax1.legend(handles[::-1], labels[::-1], loc='lower left', fontsize=20)

ax1.tick_params(axis='both', which='major', labelsize=14)
ax1.tick_params(axis='both', which='minor', labelsize=14)
ax2.tick_params(axis='both', which='major', labelsize=14)
ax2.tick_params(axis='both', which='minor', labelsize=14)
ax3.tick_params(axis='both', which='major', labelsize=14)
ax3.tick_params(axis='both', which='minor', labelsize=14)
#ax3.ticklabel_format(axis="y", style="sci", scilimits=(0,0))
#ax3.yaxis.get_offset_text().set_fontsize(14)
#ax3.yaxis.offsetText.set_visible(False)
#ax3.text(5.3, 0.00093, '$\\times\\mathdefault{10^{3}}\\mathdefault{}$', fontsize=14)
plt.xlim(min(fgrids), max(fgrids))
ax1.grid(True, which='both', ls='dotted', linewidth='0.8', alpha=.8)
ax2.grid(True, which='both', ls='dotted', linewidth='0.8', alpha=.8)
ax3.grid(True, which='both', ls='dotted', linewidth='0.8', alpha=.8)

ax1.set_title(r"{$\bf \texttt{IMRPhenomD}$}", fontsize=25)

plt.savefig(os.path.join('Plots_WF_comparison', 'WF_Comparison_IMRPhenomD.pdf'))

plt.show()
```
